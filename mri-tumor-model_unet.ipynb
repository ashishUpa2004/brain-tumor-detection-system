```python

import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

# ── 1. Imports ───────────────────────────────────────────────
import numpy as np
import random
import cv2
import h5py
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from collections import Counter

from sklearn.utils import shuffle
from sklearn.utils import class_weight as sklearn_class_weight
from sklearn.preprocessing import LabelEncoder, label_binarize
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Dense, Dropout,
    Conv2D, MaxPooling2D, Conv2DTranspose,
    concatenate, BatchNormalization, Activation,
    GlobalAveragePooling2D, SpatialDropout2D
)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import VGG16
from tensorflow.keras.callbacks import (
    EarlyStopping, ReduceLROnPlateau,
    ModelCheckpoint, LambdaCallback
)
from tensorflow.keras.regularizers import l2

print('TensorFlow version:', tf.__version__)
print('Available devices :', tf.config.list_physical_devices())

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

AUTOTUNE   = tf.data.AUTOTUNE
BATCH_SIZE = 20
IMAGE_SIZE = 128
EPOCHS     = 50


#          PART A — VGG16 CLASSIFICATION                

# ── 2. Find Dataset ──────────────────────────────────────────
def find_clf_dirs(base='/kaggle/input'):
    train_dir = test_dir = None
    for root, dirs, files in os.walk(base):
        folder = os.path.basename(root).lower()
        if folder == 'training' and train_dir is None:
            if any(os.path.isdir(os.path.join(root, d)) for d in os.listdir(root)):
                train_dir = root
        if folder == 'testing' and test_dir is None:
            if any(os.path.isdir(os.path.join(root, d)) for d in os.listdir(root)):
                test_dir = root
    return train_dir, test_dir

train_dir, test_dir = find_clf_dirs()

if train_dir is None or test_dir is None:
    for root, dirs, files in os.walk('/kaggle/input'):
        level = root.replace('/kaggle/input', '').count(os.sep)
        print('  '*level + os.path.basename(root) + '/')
    raise FileNotFoundError('Training/Testing folders not found.')

print(f'Train dir : {train_dir}')
print(f'Test  dir : {test_dir}')
print(f'Classes   : {sorted(os.listdir(train_dir))}')


# ── 3. Load Data ─────────────────────────────────────────────
train_paths, train_labels = [], []
for label in os.listdir(train_dir):
    lp = os.path.join(train_dir, label)
    if not os.path.isdir(lp):
        continue
    for img in os.listdir(lp):
        if img.lower().endswith(('.jpg', '.jpeg', '.png')):
            train_paths.append(os.path.join(lp, img))
            train_labels.append(label)

train_paths, train_labels = shuffle(train_paths, train_labels, random_state=SEED)

test_paths, test_labels = [], []
for label in os.listdir(test_dir):
    lp = os.path.join(test_dir, label)
    if not os.path.isdir(lp):
        continue
    for img in os.listdir(lp):
        if img.lower().endswith(('.jpg', '.jpeg', '.png')):
            test_paths.append(os.path.join(lp, img))
            test_labels.append(label)

test_paths, test_labels = shuffle(test_paths, test_labels, random_state=SEED)

print(f'Train: {len(train_paths)} | Test: {len(test_paths)}')
print(f'Classes: {sorted(set(train_labels))}')

counter = Counter(train_labels)
plt.figure(figsize=(8, 4))
plt.bar(counter.keys(), counter.values(), color='steelblue', edgecolor='black')
plt.title('Class Distribution'); plt.xlabel('Class'); plt.ylabel('Count')
plt.tight_layout(); plt.show()

fig, axes = plt.subplots(2, 5, figsize=(16, 7))
axes = axes.ravel()
for i, idx in enumerate(random.sample(range(len(train_paths)), 10)):
    axes[i].imshow(Image.open(train_paths[idx]).resize((IMAGE_SIZE, IMAGE_SIZE)), cmap='gray')
    axes[i].set_title(train_labels[idx], fontsize=9)
    axes[i].axis('off')
plt.suptitle('Sample Training Images', fontweight='bold')
plt.tight_layout(); plt.show()


# ── 4. Label Encoding & Class Weights ────────────────────────
le = LabelEncoder()
train_labels_enc = le.fit_transform(train_labels)
test_labels_enc  = le.transform(test_labels)
print('Label mapping:', dict(zip(le.classes_, le.transform(le.classes_))))

# FIX 1: Boosted glioma weight to 4.5
class_weight_dict = {
    list(le.classes_).index('glioma')     : 4.5,  
    list(le.classes_).index('meningioma') : 2.0,  
    list(le.classes_).index('notumor')    : 0.6,  
    list(le.classes_).index('pituitary')  : 0.8,  
}
print('Class weights:', class_weight_dict)


# ── 5. tf.data Pipeline ──────────────────────────────────────
def parse_image(path, label):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img = tf.image.resize(img, [IMAGE_SIZE, IMAGE_SIZE])
    img = tf.cast(img, tf.float32) / 255.0
    return img, label

#  FIX 3: Better augmentation for glioma shape learning
def augment_clf(img, label):
    img = tf.image.random_flip_left_right(img)
    img = tf.image.random_flip_up_down(img)
    img = tf.image.random_brightness(img, 0.2)
    img = tf.image.random_contrast(img, 0.8, 1.2)
    img = tf.image.random_saturation(img, 0.8, 1.2)
    img = tf.image.random_hue(img, 0.05)
    #  Random zoom helps glioma learn shape better
    crop = tf.random.uniform([], int(IMAGE_SIZE*0.88), IMAGE_SIZE, dtype=tf.int32)
    img  = tf.image.random_crop(img, [crop, crop, 3])
    img  = tf.image.resize(img, [IMAGE_SIZE, IMAGE_SIZE])
    return tf.clip_by_value(img, 0.0, 1.0), label

def make_clf_dataset(paths, labels, batch_size, augment_data=True, do_shuffle=True):
    ds = tf.data.Dataset.from_tensor_slices((
        tf.constant(paths), tf.constant(labels, dtype=tf.int32)
    ))
    if do_shuffle:
        ds = ds.shuffle(len(paths), seed=SEED)
    ds = ds.map(parse_image, num_parallel_calls=AUTOTUNE)
    if augment_data:
        ds = ds.map(augment_clf, num_parallel_calls=AUTOTUNE)
    return ds.batch(batch_size).prefetch(AUTOTUNE)

train_dataset = make_clf_dataset(train_paths, train_labels_enc, BATCH_SIZE)
val_dataset   = make_clf_dataset(test_paths, test_labels_enc, BATCH_SIZE,
                                  augment_data=False, do_shuffle=False)
print('Classification datasets ready.')


# ── 6. VGG16 Model ───────────────────────────────────────────
REG        = l2(1e-4)
base_model = VGG16(input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3),
                   include_top=False, weights='imagenet')
for layer in base_model.layers:
    layer.trainable = False

inputs  = Input(shape=(IMAGE_SIZE, IMAGE_SIZE, 3))
x       = base_model(inputs, training=False)
x       = GlobalAveragePooling2D()(x)
x       = BatchNormalization()(x)
x       = Dropout(0.4)(x)
x       = Dense(512, activation='relu', kernel_regularizer=REG)(x)
x       = BatchNormalization()(x)
x       = Dropout(0.3)(x)
x       = Dense(256, activation='relu', kernel_regularizer=REG)(x)
x       = Dropout(0.2)(x)
outputs = Dense(len(le.classes_), activation='softmax')(x)
clf_model = Model(inputs, outputs, name='VGG16_Classifier')


# ── 7. Phase 1 — Train Head Only ─────────────────────────────
print('\n=== Phase 1: Train classifier head ===')
clf_model.compile(optimizer=Adam(1e-3),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

h1 = clf_model.fit(
    train_dataset, epochs=25,
    validation_data=val_dataset,
    class_weight=class_weight_dict,
    callbacks=[
        EarlyStopping(monitor='val_loss', patience=6,
                      restore_best_weights=True),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5,
                          patience=2, min_lr=1e-6, verbose=1)
    ]
)


# ── 8. Phase 2 — Fine-tune Last 12 Layers ────────────────────
print('\n=== Phase 2: Fine-tune last 12 VGG16 layers ===')
base_model.trainable = True


for layer in base_model.layers[:-12]:
    layer.trainable = False

clf_model.compile(optimizer=Adam(1e-5),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

h2 = clf_model.fit(
    train_dataset, epochs=EPOCHS,
    validation_data=val_dataset,
    class_weight=class_weight_dict,
    callbacks=[
       
        EarlyStopping(monitor='val_loss', patience=12,
                      restore_best_weights=True, verbose=1),
        ReduceLROnPlateau(monitor='val_loss', factor=0.3,
                          patience=3, min_lr=1e-8, verbose=1),
        ModelCheckpoint('/kaggle/working/best_vgg16.keras',
                        monitor='val_accuracy', save_best_only=True,
                        mode='max', verbose=1)
    ]
)


# ── 9. Classification Results ────────────────────────────────
acc      = h1.history['accuracy']     + h2.history['accuracy']
val_acc  = h1.history['val_accuracy'] + h2.history['val_accuracy']
loss     = h1.history['loss']         + h2.history['loss']
val_loss = h1.history['val_loss']     + h2.history['val_loss']
ep       = range(1, len(acc)+1)
p1_end   = len(h1.history['accuracy'])

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(ep, acc,     label='Train')
axes[0].plot(ep, val_acc, label='Val')
axes[0].axvline(p1_end, color='gray', linestyle='--', label='Phase 1→2')
axes[0].set_title('Accuracy'); axes[0].legend()
axes[1].plot(ep, loss,     label='Train')
axes[1].plot(ep, val_loss, label='Val')
axes[1].axvline(p1_end, color='gray', linestyle='--', label='Phase 1→2')
axes[1].set_title('Loss'); axes[1].legend()
plt.suptitle('VGG16 Training History', fontweight='bold')
plt.tight_layout(); plt.show()

gap = val_loss[-1] - loss[-1]
if gap > 0.3:            print(f'  Overfitting!   gap={gap:.3f}')
elif val_acc[-1] < 0.70: print(f'  Underfitting!  val_acc={val_acc[-1]:.2%}')
else:                    print(f' Healthy model!  val_acc={val_acc[-1]:.2%}')

y_pred         = clf_model.predict(val_dataset)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true         = np.concatenate([y for _, y in val_dataset], axis=0)

print('\nClassification Report:')
print(classification_report(y_true, y_pred_classes, target_names=le.classes_))

# Per-class accuracy chart
class_acc = {}
for i, cls in enumerate(le.classes_):
    mask = y_true == i
    class_acc[cls] = (y_pred_classes[mask] == y_true[mask]).mean() * 100

plt.figure(figsize=(8, 5))
colors = ['red' if v < 75 else 'green' for v in class_acc.values()]
plt.bar(class_acc.keys(), class_acc.values(), color=colors, edgecolor='black')
plt.axhline(y=75, color='orange', linestyle='--', label='75% threshold')
plt.title('Per-Class Accuracy', fontweight='bold')
plt.ylabel('Accuracy %'); plt.ylim(0, 100); plt.legend()
for i, (cls, val) in enumerate(class_acc.items()):
    plt.text(i, val+1, f'{val:.1f}%', ha='center', fontweight='bold')
plt.tight_layout(); plt.show()

cm = confusion_matrix(y_true, y_pred_classes)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted'); plt.ylabel('Actual')
plt.title('Confusion Matrix — VGG16')
plt.tight_layout(); plt.show()

n_classes  = len(le.classes_)
y_true_bin = label_binarize(y_true, classes=np.arange(n_classes))
fpr, tpr, roc_auc = {}, {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_pred[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
plt.figure(figsize=(10, 8))
for i, color in zip(range(n_classes), plt.cm.tab10(np.linspace(0, 1, n_classes))):
    plt.plot(fpr[i], tpr[i], color=color,
             label=f'{le.classes_[i]} (AUC={roc_auc[i]:.2f})')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('FPR'); plt.ylabel('TPR')
plt.title('ROC Curves — VGG16')
plt.legend(loc='lower right'); plt.tight_layout(); plt.show()

print('\n' + '='*50)
print('      PER-CLASS ACCURACY SUMMARY')
print('='*50)
for cls, acc_val in class_acc.items():
    status = 'good' if acc_val >= 75 else 'not good'
    bar    = '█' * int(acc_val / 2.5)
    print(f'  {status} {cls:15s}: {acc_val:5.1f}%  {bar}')
print('='*50)



#      PART B — U-Net SEGMENTATION (BraTS2020 .h5)        


# ── 10. Load BraTS2020 ────────────────────────────────────────
BRATS_H5_DIR = '/kaggle/input/datasets/awsaf49/brats2020-training-data/BraTS2020_training_data/content/data'

h5_files = [f for f in os.listdir(BRATS_H5_DIR) if f.endswith('.h5')]
print(f'\nTotal .h5 files: {len(h5_files)}')

with h5py.File(os.path.join(BRATS_H5_DIR, h5_files[0]), 'r') as f:
    print(f'Keys: {list(f.keys())}')
    for key in f.keys():
        print(f'  {key} → shape: {f[key].shape}')


# ── 11. Read Slices ───────────────────────────────────────────
print('\nLoading slices...')
MAX_SLICES  = 8000
brats_imgs  = []
brats_masks = []
random.shuffle(h5_files)

for fname in h5_files:
    if len(brats_imgs) >= MAX_SLICES:
        break
    try:
        with h5py.File(os.path.join(BRATS_H5_DIR, fname), 'r') as f:
            img_data  = f['image'][()]
            mask_data = f['mask'][()]

            if img_data.ndim == 3:
                flair = img_data[..., 0] if img_data.shape[-1] == 4 else img_data[0]
            else:
                flair = img_data

            if mask_data.ndim == 3:
                mask_bin = mask_data.sum(axis=-1) if mask_data.shape[-1] in [1,3,4] \
                           else mask_data.sum(axis=0)
            else:
                mask_bin = mask_data

            mask_bin = (mask_bin > 0).astype(np.float32)
            if mask_bin.sum() < 100:
                continue

            flair = flair.astype(np.float32)
            if flair.max() > 0:
                flair /= flair.max()

            flair    = cv2.resize(flair,    (IMAGE_SIZE, IMAGE_SIZE))
            mask_bin = cv2.resize(mask_bin, (IMAGE_SIZE, IMAGE_SIZE),
                                  interpolation=cv2.INTER_NEAREST)

            brats_imgs.append(np.stack([flair]*3, axis=-1).astype(np.float32))
            brats_masks.append(mask_bin[..., np.newaxis].astype(np.float32))
    except Exception:
        continue

brats_imgs  = np.array(brats_imgs,  dtype=np.float32)
brats_masks = np.array(brats_masks, dtype=np.float32)
print(f'Slices: {len(brats_imgs)} | Tumor ratio: {brats_masks.mean():.3f}')

if len(brats_imgs) == 0:
    raise ValueError('No slices loaded!')


# ── 12. Visualise ────────────────────────────────────────────
n_show = min(4, len(brats_imgs))
fig, axes = plt.subplots(3, n_show, figsize=(16, 12))
for col, idx in enumerate(random.sample(range(len(brats_imgs)), n_show)):
    img  = brats_imgs[idx]
    mask = brats_masks[idx, ..., 0]
    overlay = img.copy()
    overlay[..., 0] = np.clip(overlay[..., 0] + 0.45*mask, 0, 1)
    axes[0,col].imshow(img[...,0], cmap='gray'); axes[0,col].set_title('FLAIR');   axes[0,col].axis('off')
    axes[1,col].imshow(mask, cmap='hot');        axes[1,col].set_title('GT Mask'); axes[1,col].axis('off')
    axes[2,col].imshow(overlay);                 axes[2,col].set_title('Overlay'); axes[2,col].axis('off')
plt.suptitle('BraTS2020 Samples', fontweight='bold')
plt.tight_layout(); plt.show()


# ── 13. Split ────────────────────────────────────────────────
idx_perm        = np.random.permutation(len(brats_imgs))
split           = int(0.85 * len(brats_imgs))
seg_train_imgs  = brats_imgs[idx_perm[:split]]
seg_train_masks = brats_masks[idx_perm[:split]]
seg_val_imgs    = brats_imgs[idx_perm[split:]]
seg_val_masks   = brats_masks[idx_perm[split:]]
print(f'Seg train: {len(seg_train_imgs)} | Seg val: {len(seg_val_imgs)}')


# ── 14. Segmentation Dataset ──────────────────────────────────
def make_seg_dataset(images, masks, batch_size,
                      augment_data=True, do_shuffle=True):
    ds = tf.data.Dataset.from_tensor_slices((images, masks))
    if do_shuffle:
        ds = ds.shuffle(len(images), seed=SEED)

    def aug(img, mask):
        combined = tf.concat([img, mask], axis=-1)
        combined = tf.image.random_flip_left_right(combined)
        combined = tf.image.random_flip_up_down(combined)
        k        = tf.random.uniform([], 0, 4, dtype=tf.int32)
        combined = tf.image.rot90(combined, k)
        crop     = tf.random.uniform([], int(IMAGE_SIZE*0.85),
                                     IMAGE_SIZE, dtype=tf.int32)
        combined = tf.image.random_crop(combined, [crop, crop, 4])
        combined = tf.image.resize(combined, [IMAGE_SIZE, IMAGE_SIZE])
        img      = tf.image.random_brightness(combined[..., :3], 0.15)
        mask     = combined[..., 3:]
        return tf.clip_by_value(img, 0.0, 1.0), mask

    if augment_data:
        ds = ds.map(aug, num_parallel_calls=AUTOTUNE)
    return ds.batch(batch_size).prefetch(AUTOTUNE)

seg_train_ds = make_seg_dataset(seg_train_imgs, seg_train_masks, BATCH_SIZE)
seg_val_ds   = make_seg_dataset(seg_val_imgs,   seg_val_masks,   BATCH_SIZE,
                                 augment_data=False, do_shuffle=False)
print('Segmentation datasets ready.')


# ── 15. U-Net Architecture ────────────────────────────────────
SEG_REG = l2(1e-5)

def conv_block(x, filters):
    x = Conv2D(filters, 3, padding='same', kernel_regularizer=SEG_REG)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Conv2D(filters, 3, padding='same', kernel_regularizer=SEG_REG)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    return x

def encoder_block(x, filters):
    skip = conv_block(x, filters)
    pool = MaxPooling2D((2, 2))(skip)
    pool = SpatialDropout2D(0.2)(pool)
    return skip, pool

def decoder_block(x, skip, filters):
    x = Conv2DTranspose(filters, (2,2), strides=2, padding='same')(x)
    x = concatenate([x, skip])
    x = Dropout(0.2)(x)
    return conv_block(x, filters)

def build_unet(input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3)):
    inputs = Input(input_shape)
    s1, p1 = encoder_block(inputs, 64)
    s2, p2 = encoder_block(p1, 128)
    s3, p3 = encoder_block(p2, 256)
    s4, p4 = encoder_block(p3, 512)
    b      = conv_block(p4, 1024)
    b      = Dropout(0.4)(b)
    d1     = decoder_block(b,  s4, 512)
    d2     = decoder_block(d1, s3, 256)
    d3     = decoder_block(d2, s2, 128)
    d4     = decoder_block(d3, s1,  64)
    out    = Conv2D(1, (1,1), activation='sigmoid', padding='same')(d4)
    return Model(inputs, out, name='UNet_BraTS2020')

unet_model = build_unet()
unet_model.summary()


# ── 16. Compile ───────────────────────────────────────────────
def dice_coefficient(y_true, y_pred, smooth=1e-6):
    y_true_f     = tf.reshape(y_true, [-1])
    y_pred_f     = tf.reshape(y_pred, [-1])
    intersection = tf.reduce_sum(y_true_f * y_pred_f)
    return (2.*intersection + smooth) / (
        tf.reduce_sum(y_true_f) + tf.reduce_sum(y_pred_f) + smooth)

def dice_loss(y_true, y_pred):
    return 1.0 - dice_coefficient(y_true, y_pred)

def combined_loss(y_true, y_pred):
    return tf.keras.losses.binary_crossentropy(y_true, y_pred) + dice_loss(y_true, y_pred)

def iou_metric(y_true, y_pred, smooth=1e-6):
    y_bin        = tf.cast(y_pred > 0.5, tf.float32)
    intersection = tf.reduce_sum(y_true * y_bin)
    union        = tf.reduce_sum(y_true) + tf.reduce_sum(y_bin) - intersection
    return (intersection + smooth) / (union + smooth)

unet_model.compile(optimizer=Adam(1e-4),
                   loss=combined_loss,
                   metrics=[dice_coefficient, iou_metric])
print('U-Net compiled.')


# ── 17. Train U-Net ───────────────────────────────────────────
unet_callbacks = [
    EarlyStopping(monitor='val_dice_coefficient', patience=10,
                  restore_best_weights=True, mode='max', verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5,
                      patience=4, min_lr=1e-8, verbose=1),
    ModelCheckpoint('/kaggle/working/best_unet_brats.keras',
                    monitor='val_dice_coefficient', save_best_only=True,
                    mode='max', verbose=1),
    LambdaCallback(on_epoch_end=lambda epoch, logs: print(
        f'   Overfitting! Train={logs["dice_coefficient"]:.3f} '
        f'Val={logs["val_dice_coefficient"]:.3f}'
    ) if logs.get("dice_coefficient",0) - logs.get("val_dice_coefficient",0) > 0.15
      else None)
]

unet_history = unet_model.fit(
    seg_train_ds, epochs=50,
    validation_data=seg_val_ds,
    callbacks=unet_callbacks
)


# ── 18. U-Net Curves ──────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for ax, (m, vm), title in zip(
    axes,
    [('loss','val_loss'),
     ('dice_coefficient','val_dice_coefficient'),
     ('iou_metric','val_iou_metric')],
    ['Loss', 'Dice Coefficient', 'IoU']
):
    tv = unet_history.history[m]
    vv = unet_history.history[vm]
    ax.plot(tv, label='Train')
    ax.plot(vv, label='Val')
    ax.fill_between(range(len(tv)), tv, vv, alpha=0.1, color='red', label='Gap')
    ax.set_title(title); ax.legend()
plt.suptitle('U-Net Training — BraTS2020', fontweight='bold')
plt.tight_layout(); plt.show()

td = unet_history.history['dice_coefficient'][-1]
vd = unet_history.history['val_dice_coefficient'][-1]
print(f'\n{"="*45}')
print(f'  Train Dice : {td:.4f}')
print(f'  Val   Dice : {vd:.4f}')
print(f'  Gap        : {td-vd:.4f}')
if   td-vd > 0.15: print('   OVERFITTING')
elif vd < 0.50:    print('   UNDERFITTING')
else:              print('   Well-fitted!')
print(f'{"="*45}')


# ── 19. Visualise Predictions ────────────────────────────────
def visualize_seg(images, masks, n=6):
    n = min(n, len(images))
    indices = random.sample(range(len(images)), n)
    fig, axes = plt.subplots(n, 4, figsize=(16, n*4))
    if n == 1: axes = axes[np.newaxis, :]
    for col, title in enumerate(['FLAIR MRI','Ground Truth','Predicted','Overlay']):
        axes[0,col].set_title(title, fontsize=11, fontweight='bold')
    for row, idx in enumerate(indices):
        img       = images[idx]
        true_mask = masks[idx, ..., 0]
        pred_mask = unet_model.predict(img[np.newaxis,...], verbose=0)[0,...,0]
        pred_bin  = (pred_mask > 0.5).astype(np.float32)
        overlay   = img.copy()
        overlay[...,0] = np.clip(overlay[...,0] + 0.4*pred_bin, 0, 1)
        dice_val  = (2*np.sum(true_mask*pred_bin)+1e-6) / (
                     np.sum(true_mask)+np.sum(pred_bin)+1e-6)
        axes[row,0].imshow(img[...,0], cmap='gray')
        axes[row,1].imshow(true_mask,  cmap='hot')
        axes[row,2].imshow(pred_mask,  cmap='hot')
        axes[row,2].set_title(f'Dice={dice_val:.2f}', fontsize=9)
        axes[row,3].imshow(overlay)
        for col in range(4): axes[row,col].axis('off')
    plt.suptitle('U-Net — BraTS2020 Results', fontsize=14, fontweight='bold')
    plt.tight_layout(); plt.show()

visualize_seg(seg_val_imgs, seg_val_masks, n=6)

print('\nFinal U-Net evaluation:')
res = unet_model.evaluate(seg_val_ds, verbose=1)
print(f'Loss={res[0]:.4f} | Dice={res[1]:.4f} | IoU={res[2]:.4f}')


# ── 20. Combined Predict ─────────────────────────────────────
def predict_full(image_path=None, brats_img=None):
    if image_path is not None:
        img_np = np.array(Image.open(image_path).resize((IMAGE_SIZE, IMAGE_SIZE)),
                          dtype=np.float32) / 255.0
        if img_np.ndim == 2:
            img_np = np.stack([img_np]*3, axis=-1)
        elif img_np.shape[-1] == 4:
            img_np = img_np[..., :3]
    else:
        img_np = brats_img

    img_batch  = img_np[np.newaxis,...]
    clf_pred   = clf_model.predict(img_batch, verbose=0)[0]
    pred_class = le.inverse_transform([np.argmax(clf_pred)])[0]
    confidence = clf_pred.max() * 100
    seg_pred   = unet_model.predict(img_batch, verbose=0)[0,...,0]
    seg_bin    = (seg_pred > 0.5).astype(np.float32)
    overlay    = img_np.copy()
    overlay[...,0] = np.clip(overlay[...,0] + 0.5*seg_bin, 0, 1)

    fig, axes = plt.subplots(1, 3, figsize=(15,5))
    axes[0].imshow(img_np[...,0], cmap='gray'); axes[0].set_title('Input MRI');    axes[0].axis('off')
    axes[1].imshow(seg_pred, cmap='hot');       axes[1].set_title('Segmentation'); axes[1].axis('off')
    axes[2].imshow(overlay);                    axes[2].set_title('Overlay');      axes[2].axis('off')
    plt.suptitle(f'Class: {pred_class.upper()}  |  Confidence: {confidence:.2f}%',
                 fontsize=13, fontweight='bold',
                 color='red' if pred_class != 'notumor' else 'green')
    plt.tight_layout(); plt.show()

    print(f'\nPredicted: {pred_class.upper()} ({confidence:.2f}%)')
    for cls, prob in zip(le.classes_, clf_pred):
        marker = ' ← PREDICTED' if cls == pred_class else ''
        print(f'  {cls:15s}: {prob*100:5.2f}%  {"█"*int(prob*40)}{marker}')
    return pred_class, confidence

print('=== BraTS2020 sample ===')
predict_full(brats_img=seg_val_imgs[0])
print('=== MRI classification sample ===')
predict_full(image_path=random.choice(test_paths))


# ── 21. Glioma & Meningioma Accuracy ─────────────────────────
print('\n' + '='*50)
print('    GLIOMA & MENINGIOMA ACCURACY')
print('='*50)
for tumor_type in ['glioma', 'meningioma']:
    images  = [p for p, l in zip(test_paths, test_labels) if l == tumor_type]
    correct = 0
    for img_path in images:
        img_np = np.array(Image.open(img_path).resize((IMAGE_SIZE, IMAGE_SIZE)),
                          dtype=np.float32) / 255.0
        if img_np.ndim == 2:
            img_np = np.stack([img_np]*3, axis=-1)
        pred       = clf_model.predict(img_np[np.newaxis,...], verbose=0)[0]
        pred_class = le.inverse_transform([np.argmax(pred)])[0]
        if pred_class == tumor_type:
            correct += 1
    accuracy = correct / len(images) * 100
    status   = 'good' if accuracy >= 75 else 'not good'
    print(f'  {status} {tumor_type.upper():15s}: {correct}/{len(images)}  ({accuracy:.1f}%)')
print('='*50)


# ── 22. Save Models ───────────────────────────────────────────
clf_model.save('/kaggle/working/MRI_VGG16_classifier.h5')
unet_model.save('/kaggle/working/MRI_UNet_BraTS2020.h5')
print('\nModels saved!')
for f in os.listdir('/kaggle/working/'):
    size = os.path.getsize(f'/kaggle/working/{f}') / 1e6
    print(f'  {f}  ({size:.1f} MB)')
```
